```python
import pandas as pd
import numpy as np


app_data = pd.read_csv('application_data.csv')
app_data.head()

```

```python
prev_app = pd.read_csv('previous_application.csv')
#print(prev_app.columns)
prev_app
```

```python
null_counts = prev_app.isnull().sum()
print(null_counts)
```

```python
mean_value = prev_app['AMT_DOWN_PAYMENT'].mean()
prev_app['AMT_DOWN_PAYMENT'] = prev_app['AMT_DOWN_PAYMENT'].fillna(mean_value)
prev_app
```

```python
import pandas as pd

# Assuming you have a DataFrame named 'prev_app' and you want to remove multiple columns
columns_to_remove = ['DAYS_TERMINATION', 'DAYS_LAST_DUE', 'DAYS_LAST_DUE_1ST_VERSION', 'DAYS_FIRST_DUE',
                     'DAYS_FIRST_DRAWING', 'WEEKDAY_APPR_PROCESS_START', 'HOUR_APPR_PROCESS_START',
                     'NFLAG_LAST_APPL_IN_DAY', 'AMT_GOODS_PRICE', 'RATE_DOWN_PAYMENT',
                     'RATE_INTEREST_PRIMARY', 'RATE_INTEREST_PRIVILEGED', 'DAYS_DECISION',
                     'NAME_PAYMENT_TYPE', 'NAME_TYPE_SUITE', 'NAME_CLIENT_TYPE', 'NAME_GOODS_CATEGORY',
                     'NAME_PORTFOLIO', 'NAME_PRODUCT_TYPE']

# Filter out the column names that exist in the DataFrame
columns_to_remove_existing = [col for col in columns_to_remove if col in prev_app.columns]

# Drop the existing columns from the DataFrame
prev_app.drop(columns_to_remove_existing, axis=1, inplace=True)

prev_app
```

```python
prev_app.dropna(axis=1, inplace=True)
prev_app
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'prev_app' and the contract type column is named 'NAME_CONTRACT_TYPE'
sns.set(style="darkgrid")
plt.figure(figsize=(8, 6))
sns.countplot(data=prev_app, x='NAME_CONTRACT_TYPE')
plt.title('Contract Type Distribution')
plt.xlabel('Contract Type')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Convert AMT_APPLICATION to lakhs
prev_app['AMT_APPLICATION_LAKHS'] = prev_app['AMT_APPLICATION'] / 100000

# Separate values in lakhs and thousands
amt_lakhs = prev_app[prev_app['AMT_APPLICATION_LAKHS'] >= 1]['AMT_APPLICATION_LAKHS']
amt_thousands = prev_app[prev_app['AMT_APPLICATION_LAKHS'] < 1]['AMT_APPLICATION_LAKHS'] * 1000

plt.figure(figsize=(10, 6))

# Plot values in lakhs
plt.hist(amt_lakhs, bins=30, edgecolor='black', alpha=0.5, label='Lakhs')

# Plot values in thousands
plt.hist(amt_thousands, bins=30, edgecolor='black', alpha=0.5, label='Thousands')

plt.title('Histogram: AMT_APPLICATION')
plt.xlabel('AMT_APPLICATION')
plt.ylabel('Frequency')

plt.legend()

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Plot NAME_CONTRACT_STATUS
plt.subplot(1, 2, 1)
sns.countplot(data=prev_app, x='NAME_CONTRACT_STATUS')
plt.title('Contract Status')

# Plot CODE_REJECT_REASON
plt.subplot(1, 2, 2)
sns.countplot(data=prev_app, x='CODE_REJECT_REASON')
plt.title('Rejection Reason')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count the frequency of each channel type
channel_counts = prev_app['CHANNEL_TYPE'].value_counts()

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(channel_counts.index, channel_counts.values)
plt.xlabel('Channel Type')
plt.ylabel('Frequency')
plt.title('Distribution of Channel Types')
plt.xticks(rotation=90)
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Calculate the frequency of each category in NAME_SELLER_INDUSTRY
seller_industry_counts = prev_app['NAME_SELLER_INDUSTRY'].value_counts()

# Calculate the frequency of each category in NAME_YIELD_GROUP
yield_group_counts = prev_app['NAME_YIELD_GROUP'].value_counts()

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Plot the grouped bar plot for NAME_SELLER_INDUSTRY
ax1.bar(np.arange(len(seller_industry_counts)), seller_industry_counts.values)
ax1.set_xticks(np.arange(len(seller_industry_counts)))
ax1.set_xticklabels(seller_industry_counts.index, rotation=90)
ax1.set_xlabel('Seller Industry')
ax1.set_ylabel('Frequency')
ax1.set_title('Distribution of Seller Industry')

# Plot the grouped bar plot for NAME_YIELD_GROUP
ax2.bar(np.arange(len(yield_group_counts)), yield_group_counts.values)
ax2.set_xticks(np.arange(len(yield_group_counts)))
ax2.set_xticklabels(yield_group_counts.index, rotation=90)
ax2.set_xlabel('Yield Group')
ax2.set_ylabel('Frequency')
ax2.set_title('Distribution of Yield Group')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Convert values to lakhs
amt_application_lakhs = prev_app['AMT_APPLICATION'] / 100000
amt_down_payment_lakhs = prev_app['AMT_DOWN_PAYMENT'] / 100000

# Plot scatter plot
plt.scatter(amt_application_lakhs, amt_down_payment_lakhs)

# Set labels and title
plt.xlabel('AMT_APPLICATION (in lakhs)')
plt.ylabel('AMOUNT_DOWN_PAYMENT (in lakhs)')
plt.title('Scatter Plot: AMT_APPLICATION vs AMOUNT_DOWN_PAYMENT')

# Format tick labels
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: '{:.0f}'.format(x)))
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: '{:.0f}'.format(x)))

# Show the plot
plt.show()
```
